```python
def slink(name):
    table_open =open('C:/Users/yhgil/Desktop/Backup/area/'+name+'.txt', encoding = 'utf-8')
    link1=[]
    lines = table_open.readlines()
    for line in lines:    
        line_list = line.replace('\n','')
        link1.append(line_list)
    table_open.close()
    return link1
```

```python
def getrate(t):
    url = 'https://www.mypetplus.co.kr'
   
    c = [] 
    for urll in t:
        d=-1
        driver.get(url+urll)
        driver.implicitly_wait(5)
        html = driver.page_source

        r = bs(html, 'html.parser')

        contents = r.find_all("div", {"class" : "star-bx"})

        for i in contents :
            a=re.findall("\d+\.\d+", i.text)
            b=re.findall("\d+", i.text)
            if a !=[]:
                d=float(a[0])
            elif b!=[]: 
                d=float(b[0])
        if d==-1 :
            c.append('none')
        else :
            c.append(d)
    return c
```

```python
def getcomment(t) :
    url = 'https://www.mypetplus.co.kr'
    blist=[]
    contents = driver.find_elements_by_xpath('//*[@id="dPostScriptList"]/div[2]') 
    
    for urll in t:
        driver.get(url+urll)
        driver.implicitly_wait(5)
        contents = driver.find_elements_by_xpath('//*[@id="dPostScriptList"]/div[2]')
        for content in contents:
            a=content.text.split('**')
            del a[0]
            blist.append(a)
    return blist
```

```python
def getcommentcount(t) :
    url = 'https://www.mypetplus.co.kr'
    blist=[]
    contents = driver.find_elements_by_xpath('//*[@id="dPostScriptList"]/div[2]') 
    
    for urll in t:
        driver.get(url+urll)
        driver.implicitly_wait(5)
        contents = driver.find_elements_by_xpath('//*[@id="dPostScriptList"]/div[2]')
        for content in contents:
            a=content.text.split('**')
            del a[0]
            b=len(a)
            blist.append(b)
    return blist
```

```python
def getname(t):
    url = 'https://www.mypetplus.co.kr'
    clist=[]
    for urll in t:
        driver.get(url+urll)
    
        html = driver.page_source

        r = bs(html, 'html.parser')

        contents = r.find_all("h2", {"class" : "tit"})
        
        for i in contents :
            blist=[0,0]
            a=i.text
            alist=a.split('-')
            blist[1]=alist[0].strip()
            blist[0]=alist[1].strip()
            clist.append(blist)
    return clist
```
